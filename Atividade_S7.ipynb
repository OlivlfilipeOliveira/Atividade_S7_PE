```
#Importando bibliotecas necessárias para realização das questões
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Exercício 1: 

```
#Lendo conjunto de dados weight_chart.txt vindo de um repositório do github
df = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/weight_chart.txt', delimiter='\t')

#plotando gráfico de ponto e linha
plt.figure(figsize=(8, 6))
plt.plot(df['Age'], df['Weight'], color='k', marker='.', linestyle='solid',  mfc='none', mec='k', markersize=18, markerfacecolor='white')
plt.xlabel('Idade (meses)')
plt.ylabel('Peso (Kg)')
plt.title('A relação entre idade e crescimento de um bebê')
plt.show()
```

# Exercício 3:

```
#gerando um conjunto de dados aleatório com 10000 elementos
dados = np.random.normal(size=10000)

#Cocatenando um conjunto que gerado acima com ele mesmo somado 4 unidades a cada elemento
dados = np.concatenate((dados, dados+4))

plt.figure(figsize=(8,5))
#abaixo é utilizado a biblioteca saeborn para plotar o histograma
#a biblioteca saeborn permite usar algumas propriedades específicas
sns.histplot(data=dados, bins=100, color='black', fill=False) 
plt.title('Histograma de Distribuição Mista')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

```
#lendo conjunto de dados male_female_counts.txt de um repositório do github
df = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/male_female_counts.txt', delimiter='\t')

#criando uma lista com as coras a serem inseridas no gráfico
cores = ['red', 'orange', 'yellow', 'green', 'aquamarine', 'turquoise', 'blue', 'navy', 'purple', 'pink']

#plotando gráfico de barras
plt.figure(figsize=(12,6))
plt.bar(df['Sample'], df['Count'], color=cores)
plt.show()
```

# Exercício 5:

```
#lendo conjunto de dados up_down_expression.txt do repositório do github
df = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/up_down_expression.txt', delimiter='\t')

#plotando o gráfico
plt.figure(figsize=(7,6))
#iterando o conjunto de dados abaixo para conseguir mudar as cores dos dados plotados no gráfico
for x, y, state in zip(df['Condition1'], df['Condition2'], df['State']):
  if(state == 'down'):
    plt.scatter(x, y, s=3, color='blue')
  elif(state == 'up'):
    plt.scatter(x, y, s=3, color='red')
  else:
    plt.scatter(x, y, s=3, color='grey')
    
plt.xlabel('Condition1')
plt.ylabel('Condition2')
plt.show()
```

# Exercício 6: 

```
#lendo conjunto de dados chromosome_position_data.txt do repositório do github
dt = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/chromosome_position_data.txt', delimiter='\t')

#plotando o gráfico 
plt.figure(figsize=(9,5))
plt.plot(dt['Position'], dt['Mut1'],color='red', label='Mut1')
plt.plot(dt['Position'], dt['Mut2'],color='blue', label='Mut2')
plt.plot(dt['Position'], dt['WT'],color='green', label='WT')
plt.xlabel('Posição Dentro do Cromossomo')
plt.ylabel('Valor')
plt.legend(loc=2)
plt.show()
```

# Exercício 7:

```
#Lendo conjunto de dados brain_bodyweight.txt do repositório do github
dt = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/brain_bodyweight.txt', delimiter='\t')

#Plotando gráfico
plt.figure(figsize=(8,9))
plt.errorbar(dt['Brainweight'], dt['Bodyweight '], fmt='*', label=dt['Species '] ,xerr=dt['Brainweight.SEM'], yerr=dt['Bodyweight.SEM'], color='black')
#iterando dados para conseguir mostrar nomes dos elementos 
for x, y, errox, erroy, nome in zip(dt['Brainweight'], dt['Bodyweight '], dt['Brainweight.SEM'], dt['Bodyweight.SEM'], dt['Species ']):
  plt.text(x-0.13, y-0.25, nome, fontsize=7)

plt.ylabel('Body weight')
plt.xlabel('Brian weight')
plt.show()
```
